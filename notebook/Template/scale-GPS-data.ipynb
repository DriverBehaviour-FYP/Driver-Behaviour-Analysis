```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import utm
```

```python
gps_data=[[7.2399066,80.6744066],
          [7.3169766,80.74408],
          [7.3170999,80.7440716],
          [7.2945016,80.7370033],
          [7.2944933,80.7367016]
          ]
```

**Prepare your data**

```python
coordinates = [[lat, lon] for lat, lon in gps_data]
```

**Convert the GPS coordinates to Cartesian coordinates (UTM)**

```python
utm_coordinates = []
for lat, lon in coordinates:
   [utm_east, utm_north, utm_zone, utm_letter] = utm.from_latlon(lat, lon)#[464056.36598958017, 800282.7156244798,44,'N']
   utm_coordinates.append([utm_east, utm_north])
utm_coordinates
```

**Scale the data**

```python
scaler = StandardScaler()
scaled_coordinates = scaler.fit_transform(utm_coordinates)
```

**Apply K-means clustering**

```python
k = 3  # number of clusters
kmeans = KMeans(n_clusters=k)
kmeans.fit(scaled_coordinates)
```

**Retrieve the cluster labels**

```python
cluster_labels = kmeans.labels_
cluster_labels
```
